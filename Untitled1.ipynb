```python
import os
import glob
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
import soundfile as sf
from torch.utils.data import Dataset, DataLoader
from scipy.signal import spectrogram
```

```python
class RealVoiceRestorationDataset(Dataset):
    def __init__(self, target_dir, max_samples=1000):
        # Scan local folder for the real downloaded .flac voice files
        self.file_paths = glob.glob(os.path.join(target_dir, "**/*.flac"), recursive=True)[:max_samples]
        self.segment_length = 32000  # Sets uniform length: 2 seconds of audio at 16kHz
        print(f"Found and linked {len(self.file_paths)} real human speech files for training.")

    def __len__(self):
        return len(self.file_paths)

    def __getitem__(self, idx):
        # Load the real human speech waveform array safely via soundfile
        waveform, fs = sf.read(self.file_paths[idx], dtype='float32')
        
        # Ensure uniform 2-second arrays (pad short clips, crop long ones)
        if len(waveform) < self.segment_length:
            waveform = np.pad(waveform, (0, self.segment_length - len(waveform)))
        else:
            waveform = waveform[:self.segment_length]

        # 1. Compute Perfect Clean Spectrogram via safe SciPy backend
        _, _, clean_spec = spectrogram(waveform, fs=fs, nperseg=400, noverlap=240, mode='complex')
        clean_mag = np.abs(clean_spec).astype(np.float32)

        # 2. Corrupt with Background Hiss Noise
        noisy_waveform = waveform + (np.random.randn(*waveform.shape).astype(np.float32) * 0.04)
        _, _, noisy_spec = spectrogram(noisy_waveform, fs=fs, nperseg=400, noverlap=240, mode='complex')
        noisy_mag = np.abs(noisy_spec).astype(np.float32)
        noisy_phase = np.angle(noisy_spec).astype(np.float32)

        # 3. Simulate Shrillness & Room Acoustic Distortions
        freq_bins, _ = noisy_mag.shape
        distortion_filter = np.ones_like(noisy_mag)
        
        # High-frequency boost curve (shrillness)
        shrill_curve = np.linspace(1.0, 3.5, num=freq_bins)[:, np.newaxis]
        distortion_filter *= shrill_curve
        
        # Room vocal resonances (hollow peaks and dropout holes)
        distortion_filter[30:60, :] *= 2.5    
        distortion_filter[100:130, :] *= 0.05  
        
        corrupted_mag = noisy_mag * distortion_filter

        # Package as PyTorch tensors [Channels, Height, Width]
        return (
            torch.from_numpy(corrupted_mag).unsqueeze(0), 
            torch.from_numpy(clean_mag).unsqueeze(0), 
            torch.from_numpy(noisy_phase)
        )
```

```python
# Direct the loader to your pre-extracted local data directory
data_dir = os.path.abspath("./data")
target_path = os.path.join(data_dir, "LibriSpeech", "dev-clean")

# Load up to 1000 real samples for balanced training speed
train_dataset = RealVoiceRestorationDataset(target_path, max_samples=1000)
train_loader = DataLoader(train_dataset, batch_size=4, shuffle=True, num_workers=0)
print("Real Voice Pipeline successfully engaged!")
```

```python
class AudioRestorationUNet(nn.Module):
    def __init__(self):
        super(AudioRestorationUNet, self).__init__()
        
        # Encoder Layers: Extracts vocal traits while discarding bad pixel noise
        self.enc1 = nn.Conv2d(1, 16, kernel_size=3, stride=1, padding=1)
        self.enc2 = nn.Conv2d(16, 32, kernel_size=3, stride=2, padding=1) 
        self.enc3 = nn.Conv2d(32, 64, kernel_size=3, stride=2, padding=1) 
        self.bn1 = nn.BatchNorm2d(16)
        self.bn2 = nn.BatchNorm2d(32)
        self.bn3 = nn.BatchNorm2d(64)
        self.act = nn.LeakyReLU(0.2)
        
        # Decoder Layers: Rebuilds clean audio matrix from the encoder details
        self.dec1 = nn.ConvTranspose2d(64, 32, kernel_size=3, stride=2, padding=1, output_padding=1)
        self.dec2 = nn.ConvTranspose2d(32, 16, kernel_size=3, stride=2, padding=1, output_padding=1)
        self.dec3 = nn.Conv2d(16, 1, kernel_size=3, stride=1, padding=1)
        self.bn_dec1 = nn.BatchNorm2d(32)
        self.bn_dec2 = nn.BatchNorm2d(16)
        self.final_act = nn.ReLU() # Forces outputs to be valid absolute magnitude values

    def forward(self, x):
        # Capture input dimensions [Height, Width] dynamically to avoid mismatch errors
        target_h, target_w = x.shape[2], x.shape[3]
        
        # Pass through the encoding tunnel
        e1 = self.act(self.bn1(self.enc1(x)))
        e2 = self.act(self.bn2(self.enc2(e1)))
        e3 = self.act(self.bn3(self.enc3(e2)))
        
        # Pass through the decoding tunnel
        d1 = self.act(self.bn_dec1(self.dec1(e3)))
        d2 = self.act(self.bn_dec2(self.dec2(d1)))
        out = self.final_act(self.dec3(d2))
        
        # Automatically slice away rounding errors to match targets exactly
        return out[:, :, :target_h, :target_w]

# Initialize the model blueprint
model = AudioRestorationUNet()
```

```python
# Setup loss evaluator and the weight fine-tuner
criterion = nn.L1Loss()
optimizer = optim.AdamW(model.parameters(), lr=0.001, weight_decay=1e-4)

print("Initiating Training Loop over Real Human Voice Templates...")
for epoch in range(5):
    model.train()
    running_loss = 0.0
    for corrupted_mag, clean_mag, _ in train_loader:
        optimizer.zero_grad() # Clear gradients from previous batch
        
        # 1. Forward Pass: Get the AI's cleanup attempt
        restored_outputs = model(corrupted_mag)
        
        # 2. Check the Error: Compare against the clean target vocal file
        loss = criterion(restored_outputs, clean_mag)
        
        # 3. Backpropagation: Walk backward through the model layers and tune knobs
        loss.backward()
        optimizer.step()
        
        running_loss += loss.item()
        
    print(f"Epoch [{epoch+1}/5] Complete | Real Vocal Loss: {running_loss/len(train_loader):.6f}")

# Save the trained AI brain settings to a file
print("\nSuccess! Real-world audio restoration model trained and saved.")
torch.save(model.state_dict(), "universal_audio_restoration_weights.pth")
```

```python
import os
import numpy as np
import torch
import soundfile as sf
import scipy.io.wavfile as wav
from scipy.signal import spectrogram, istft

# 1. Load the architecture and the freshly trained real-vocal weights
test_model = AudioRestorationUNet()
test_model.load_state_dict(torch.load("universal_audio_restoration_weights.pth"))
test_model.eval()  # Set to evaluation mode
print("✨ Real-World Audio Restoration Engine Loaded & Ready!")

def restore_human_voice(input_path, output_path):
    if not os.path.exists(input_path):
        print(f"❌ Error: Could not find '{input_path}' in your folder. Double-check the filename!")
        return

    # 2. Read your test file safely
    waveform, fs = sf.read(input_path, dtype='float32')
    
    # If stereo, blend it down to mono
    if len(waveform.shape) > 1:
        waveform = np.mean(waveform, axis=1)
        
    # 3. Pass to SciPy to extract the magnitude image and phase map
    nperseg, noverlap = 400, 240
    _, _, spec = spectrogram(waveform, fs=fs, nperseg=nperseg, noverlap=noverlap, mode='complex')
    corrupted_mag = np.abs(spec).astype(np.float32)
    original_phase = np.angle(spec).astype(np.float32)

    # 4. Neural Network Inference
    input_tensor = torch.from_numpy(corrupted_mag).unsqueeze(0).unsqueeze(0) # [Batch=1, Channel=1, H, W]
    with torch.no_grad():
        restored_tensor = test_model(input_tensor)
    
    restored_mag = restored_tensor.squeeze(0).squeeze(0).numpy()

    # 5. Reconstruct back into raw sound waves using the original phase map
    restored_spec = restored_mag * np.exp(1j * original_phase)
    _, restored_waveform = istft(restored_spec, fs=fs, nperseg=nperseg, noverlap=noverlap)

    # Prevent loud audio clipping blowouts
    if np.max(np.abs(restored_waveform)) > 0:
        restored_waveform = restored_waveform / np.max(np.abs(restored_waveform))
    
    # 6. Save the clean output file back to your drive
    wav.write(output_path, fs, (restored_waveform * 32767).astype(np.int16))
    print(f"🎉 Mastered audio successfully saved to disk as: {output_path}")

# =====================================================================
# EXECUTE THE TEST
# =====================================================================
# Change "my_test_voice.wav" to match your actual input filename!
restore_human_voice("test.wav", "restored_output.wav")
```
